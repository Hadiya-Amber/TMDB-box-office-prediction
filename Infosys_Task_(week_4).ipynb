## 1. Extract release_year, release_month and release_day as different column

```python
# Converting the "release_date" column to datetime format
data['release_date'] = pd.to_datetime(data['release_date'], format='%d-%m-%Y')

# Extracting the year, month and day
data['release_year'] = data['release_date'].dt.year
data['release_month'] = data['release_date'].dt.month
data['release_day'] = data['release_date'].dt.day
```

## 3. Using One hot encoding on "genres", "spoken_languages" and "production_companies" columns

#### a. For Genres

```python
# Spliting the genres into a list of genres
data['genres'] = data['genres'].apply(lambda x: x.split(','))

# Get unique genres
unique_genres = {genre for sublist in data['genres'] for genre in sublist}
print(unique_genres)
```

```python
# Create binary columns for each genre
for genre in unique_genres:
    data[genre] = data['genres'].apply(lambda x: 1 if genre in x else 0)
```

```python
data.drop(columns=['genres'], inplace=True)
data.columns
```

```python
data.head()
```

#### b. For Spoken Languages

```python
# Use str.get_dummies() to create binary columns for each spoken language
language_dummies = data['spoken_languages'].str.get_dummies(sep=',')

# Concatenate the new language columns with the original DataFrame
data = pd.concat([data, language_dummies], axis=1)
```

```python
data.columns
```

```python
data.head()
```

```python
data = data.drop(columns=['spoken_languages'])
```

```python
data
```

#### c. For Production Countries

```python
# Use str.get_dummies() to create binary columns for each production countries
companies_dummies = data['production_countries'].str.get_dummies(sep=',')

# Concatenate the new language columns with the original DataFrame
data = pd.concat([data, companies_dummies], axis=1)
```

```python
data = data.drop(columns=['production_countries'])
```

## 4. Convert Categorical values to Numerical Values

#### a. For status

```python
# Convert the status column to numerical value 1 = "Released", 0 = "Not Released"
data['status'] = data['status'].apply(lambda x: 1 if x == 'Released' else 0)
```

#### b. For original_language

```python
unique_languages = data['original_language'].unique()
print(unique_languages)
```

```python
# Convert the status column to numerical value 1 = "en", 0 = "else"
data['original_language'] = data['original_language'].apply(lambda x: 1 if x == 'en' else 0)
```
