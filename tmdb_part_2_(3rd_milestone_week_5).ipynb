```python
import pandas as pd
import numpy as np
from pycaret.regression import *
```

```python
data = pd.read_csv("/content/drive/MyDrive/Infosys Internship/latest_tmdb_dataset.csv")
data.head()
```

```python
data_test = pd.read_csv("/content/drive/MyDrive/My Datasets/pre-processed_testdata.csv")
data_test.head()
```

```python
data.shape, data_test.shape
```

##  Managing the columns

```python
columns_data = set(data.columns)
columns_data_test = set(data_test.columns)

common_columns = columns_data.intersection(columns_data_test)
only_in_column_data = columns_data.difference(columns_data_test)
only_in_column_data_test = columns_data_test.difference(columns_data)
```

```python
print("Common columns:", common_columns)
print("Columns only in df1:", only_in_column_data)
print("Columns only in df2:", only_in_column_data_test)
```

```python
for col in only_in_column_data_test:
    data[col] = 0
```

```python
data.shape, data_test.shape
```

```python
# Drop the extra 'revenue' column from data_test
data_test = data_test.drop('revenue', axis=1)
```

## Setting up the pycaret regression setup


```python
data.columns
```

```python
reg = setup(
    data = data,
    target = 'revenue',
    use_gpu = False,
    train_size = 0.8,
    normalize = True,
    remove_outliers = True,
    outliers_threshold = 0.05,
    numeric_imputation = 'mean',
    categorical_imputation = 'mode',
    ignore_features = ['id', 'homepage', 'imdb_id', 'movie_title', 'overview', 'poster_path', 'tagline', 'title', 'keywords'],
    session_id = 3000
)
```

## Finding the best 3 models to work on

```python
best = compare_models(n_select = 3)
```

```python
print(best)
```

## Using Elastic Net to train the model

```python
# Creating ElasticNet model
elastic_net = create_model('en', fold=10)
```

```python
tuned_elastic_net = tune_model(elastic_net)
```

## Plotting the Graphs

```python
# Residual plot
plot_model(tuned_elastic_net)
```

```python
plot_model(tuned_elastic_net, plot = 'error')
```

```python
plot_model(tuned_elastic_net, plot = 'feature')
```

```python
evaluate_model(elastic_net)
```

## Testing the model

```python
predict_model(tuned_elastic_net)
```

```python
test_pred = predict_model(tuned_elastic_net, data = data)
```

```python
test_pred[['id', 'revenue', 'prediction_label']].to_csv('trainingPredictions.csv')
```

```python
test_pred = predict_model(tuned_elastic_net, data = data_test)
```
