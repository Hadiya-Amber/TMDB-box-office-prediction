```python
base_url = "https://api.themoviedb.org/3/movie/"
def fetch_movie_details(movie_id):
    response = requests.get(f"{base_url}{movie_id}", headers=headers)
    if response.status_code == 200:
        movie_data = response.json()
        if movie_data.get('budget', 0) > 0:
            return {
                'id' : movie_id,
                'budget': movie_data.get('budget'),
                'genres': movie_data.get('genres'),
                'homepage': movie_data.get('homepage'),
                'imdb_id': movie_data.get('imdb_id'),
                'original_language': movie_data.get('original_language'),
                'original_title': movie_data.get('original_title'),
                'overview': movie_data.get('overview'),
                'popularity': movie_data.get('popularity'),
                'poster_path': movie_data.get('poster_path'),
                'production_companies': movie_data.get('production_companies'),
                'production_countries': movie_data.get('production_countries'),
                'release_date': movie_data.get('release_date'),
                'runtime': movie_data.get('runtime'),
                'spoken_languages': movie_data.get('spoken_languages'),
                'status': movie_data.get('status'),
                'tagline': movie_data.get('tagline'),
                'title': movie_data.get('title'),
                'keywords': fetch_movie_keywords(movie_id),
                'cast': fetch_movie_credits(movie_id, 'cast'),
                'crew': fetch_movie_credits(movie_id, 'crew'),
                'revenue': movie_data.get('revenue')
            }
    return None

def fetch_movie_keywords(movie_id):
    response = requests.get(f"{base_url}{movie_id}/keywords", headers=headers)
    if response.status_code == 200:
        keywords_data = response.json()
        return keywords_data.get('keywords', [])
    return []

def fetch_movie_credits(movie_id, credit_type):
    response = requests.get(f"{base_url}{movie_id}/credits", headers=headers)
    if response.status_code == 200:
        credits_data = response.json()
        return credits_data.get(credit_type, [])
    return []

def fetch_movies_data(movie_ids):
    movie2008Info = []
    count = 0
    
    for movie_id in movie_ids:
        if count >= 100:
            break
        movie_data = fetch_movie_details(movie_id)
        if movie_data and movie_data['budget'] > 0:
            movie2008Info.append(movie_data)
            count += 1
            
    return movie2008Info
```

```python
movie_data = fetch_movies_data(movie_ids)
movies_df = pd.DataFrame(movie_data)
movies_df.to_csv('movies2008_data.csv', index=False)
```
