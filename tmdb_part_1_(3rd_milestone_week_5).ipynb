### Extract release_year, release_month and release_day as different column

```python
# Converting the "release_date" column to datetime format
data_test['release_date'] = pd.to_datetime(data_test['release_date'])

# Extracting the year, month and day
data_test['release_year'] = data_test['release_date'].dt.year
data_test['release_month'] = data_test['release_date'].dt.month
data_test['release_day'] = data_test['release_date'].dt.day

# Dropping "release_date" coulmn
data_test = data_test.drop('release_date', axis=1)
```

### Using One hot encoding on "genres", "spoken_languages" and "production_companies" columns

#### a. For Genres

```python
# Spliting the genres into a list of genres
data_test['genres'] = data_test['genres'].apply(lambda x: x.split(','))

# Get unique genres
unique_genres = {genre for sublist in data_test['genres'] for genre in sublist}
print(unique_genres)
```

```python
# Create binary columns for each genre
for genre in unique_genres:
    data_test[genre] = data_test['genres'].apply(lambda x: 1 if genre in x else 0)
```

```python
data_test.drop(columns=['genres'], inplace=True)
data_test.columns
```

```python
data_test.head()
```

#### b. For Spoken Languages

```python
# Use str.get_dummies() to create binary columns for each spoken language
language_dummies = data_test['spoken_languages'].str.get_dummies(sep=',')

# Concatenate the new language columns with the original DataFrame
data_test = pd.concat([data_test, language_dummies], axis=1)
```

```python
data_test = data_test.drop(columns=['spoken_languages'])
```

#### c. For Production Countries

```python
# Use str.get_dummies() to create binary columns for each production countries
companies_dummies = data_test['production_countries'].str.get_dummies(sep=',')

# Concatenate the new language columns with the original DataFrame
data_test = pd.concat([data_test, companies_dummies], axis=1)
```
